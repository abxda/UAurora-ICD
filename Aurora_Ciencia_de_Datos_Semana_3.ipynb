# 4. **Gráficas de Puntos (Dot Plots)**

### **Teoría General**

Las **gráficas de puntos** o **dot plots** son un tipo de visualización en la que **cada valor individual** en un conjunto de datos es representado como un punto a lo largo de una escala numérica. Es una manera sencilla y directa de visualizar **distribuciones de datos discretos o continuos**, permitiendo observar claramente la **frecuencia** y **dispersión** de los datos.

- **Estructura**: En una gráfica de puntos, los valores se representan en una línea numérica, y los puntos se apilan verticalmente si hay valores repetidos. Esto ayuda a visualizar tanto la **frecuencia** como la **distribución** de los datos en un conjunto.
  
  Las gráficas de puntos son útiles cuando:
  
  - El tamaño del conjunto de datos es pequeño o mediano.
  - Se busca comparar distribuciones entre diferentes grupos.
  - Se desea ver cómo los datos están **agrupados** o **dispersos** en una escala.
- **Ventajas**:
  
  - Permiten observar cada punto de datos individualmente, lo que ofrece una visión clara de la distribución sin necesidad de agregación.
  - Son ideales para identificar patrones, agrupaciones o valores atípicos en conjuntos de datos pequeños o medianos.
- **Limitaciones**:
  
  - No son tan efectivas para conjuntos de datos muy grandes, ya que pueden volverse difíciles de interpretar cuando hay demasiados puntos.

### **Aplicaciones Comunes**:

Las gráficas de puntos se utilizan ampliamente en:

- **Educación**: Para representar calificaciones o resultados de pruebas.
- **Estadísticas descriptivas**: Para mostrar la distribución de variables numéricas.
- **Comparación de grupos**: Comparar las distribuciones entre diferentes categorías.

---

### **Cálculo y Creación de una Gráfica de Puntos**

Vamos a crear una gráfica de puntos utilizando el mismo conjunto de datos de calificaciones que hemos usado antes:

```python
import numpy as np
import matplotlib.pyplot as plt

# Conjunto de datos: calificaciones de un grupo de estudiantes
data = [85, 90, 92, 88, 76, 95, 89, 85,85, 78, 81, 83, 83]

# Contar la frecuencia de cada valor
values, counts = np.unique(data, return_counts=True)

# Ajustar el tamaño de la figura y la visualización
fig, ax = plt.subplots(figsize=(6, 2.5))

# Dibujar la gráfica de puntos apilados con el formato adecuado
for value, count in zip(values, counts):
    ax.plot([value] * count, list(range(count)), 'bo', ms=10, linestyle='')

# Configuraciones de formato del gráfico
for spine in ['top', 'right', 'left']:
    ax.spines[spine].set_visible(False)

ax.yaxis.set_visible(False)
ax.set_ylim(-1, max(counts))
ax.set_xticks(range(min(values), max(values)+1))

# Ajustar el tamaño de las etiquetas del eje X
ax.tick_params(axis='x', length=0, pad=8, labelsize=8)  # Ajustamos el tamaño a 8

# Título y etiquetas
ax.set_xlabel('Calificaciones', fontsize=10)
ax.set_title('Gráfica de Puntos (Dot Plot) Apilados', fontsize=12)

plt.show()
```

Las **gráficas de puntos** ofrecen una manera simple y directa de visualizar datos individuales en una escala numérica. Son especialmente útiles para conjuntos de datos pequeños o medianos, ya que permiten observar cada valor individualmente, mostrando la **frecuencia** y la **distribución** sin necesidad de agrupar los datos.

Con este tipo de visualización, es fácil identificar patrones, valores atípicos o diferencias entre grupos. Si bien tienen algunas limitaciones para conjuntos de datos muy grandes, las gráficas de puntos son una excelente herramienta cuando se busca claridad y simplicidad en la visualización de datos numéricos.

### 5.1 **Tablas de Frecuencias: Definición y Elaboración Paso a Paso**

#### **Teoría**

Una **tabla de frecuencias** es una herramienta utilizada en estadísticas descriptivas que organiza un conjunto de datos en clases o intervalos, contando la **frecuencia absoluta** (el número de veces que ocurre un valor o grupo de valores) y calculando la **frecuencia relativa** (la proporción de veces que ocurre un valor o grupo de valores con respecto al total de datos). Esta tabla puede incluir también la **frecuencia acumulada**, que es la suma de las frecuencias absolutas hasta un cierto punto.

La construcción de una tabla de frecuencias consta de los siguientes pasos:

1. **Determinar el rango o amplitud de los datos**: Se calcula la diferencia entre el valor máximo y el valor mínimo del conjunto de datos.
2. **Determinar el número de clases**: Utilizando métodos como la raíz cuadrada de \(n\) o la fórmula de Sturges, se decide cuántas clases o intervalos se usarán.
3. **Calcular el tamaño de las clases**: Se divide la amplitud de los datos entre el número de clases para obtener el tamaño de cada clase.
4. **Definir los límites de las clases**: Se determina el rango de cada clase, empezando por el valor mínimo y sumando el tamaño de clase hasta el valor máximo.
5. **Contar la frecuencia en cada clase**: Se cuenta cuántos datos caen dentro de cada clase.
6. **Calcular la frecuencia relativa**: Se calcula dividiendo la frecuencia absoluta entre el total de datos.
7. **Calcular la frecuencia acumulada**: Se obtiene sumando las frecuencias absolutas hasta cada clase.

---

### **Ejemplo en Python: Tabla de Frecuencias Paso a Paso**

Utilizaremos un conjunto de calificaciones de estudiantes para ilustrar cómo construir una tabla de frecuencias de manera automatizada.

#### **Paso 1: Determinar la amplitud de los datos**

La **amplitud de variación** es la diferencia entre el valor máximo y el valor mínimo en el conjunto de datos.

```python
import numpy as np

# Conjunto de datos: calificaciones de un grupo de estudiantes
data = [85, 90, 78, 92, 88, 76, 95, 89, 85, 90, 91, 84, 93, 87, 77, 96, 85, 90, 88, 86,
        92, 81, 89, 93, 90, 79, 85, 87, 92, 94, 83, 88, 86, 95, 91, 80, 84, 82, 94, 89,
        88, 90, 92, 85, 96, 87, 85, 93, 91, 89]

# Calcular el valor mínimo, máximo y la amplitud de variación
minimo = min(data)
maximo = max(data)
amplitud_variacion = maximo - minimo

print(f"Valor mínimo: {minimo}, Valor máximo: {maximo}, Amplitud de variación: {amplitud_variacion}")
```

**Paso 2: Determinar el número de clases**

El **número de clases** es el número de intervalos en los que queremos dividir nuestros datos para organizar la tabla. Existen varios métodos para determinar este número, pero uno de los más comunes es la **fórmula de Sturges**:

### **Fórmula de Sturges**:

$
C = 1 + 3.322 \cdot \log_{10}(n)
$

Donde:

- $C$ es el número de clases.
- $n$ es el número total de observaciones.

Esta fórmula toma en cuenta el tamaño del conjunto de datos para ajustar el número de clases. **Intuitivamente**, la fórmula de Sturges busca un equilibrio entre tener demasiadas clases (lo que podría hacer difícil interpretar los datos) y tener muy pocas clases (lo que podría ocultar detalles importantes de la distribución de los datos).

#### **Ejemplo en Python**:

```python
import math

# Número total de datos
n = len(data)

# Fórmula de Sturges para determinar el número de clases
clases_sturges = round(1 + 3.322 * math.log10(n))
print(f"Número de clases (Sturges): {clases_sturges}")
```

La **fórmula de Sturges** ha determinado que para este conjunto de datos de 50 observaciones, el número ideal de clases es **7**.

#### **Sentido Intuitivo de la Fórmula de Sturges**:

- Si tenemos pocos datos (\(n\) pequeño), la fórmula tenderá a dar un número pequeño de clases, lo que tiene sentido porque no queremos sobrecargar los datos con demasiados intervalos.
- Si tenemos muchos datos (\(n\) grande), la fórmula sugiere más clases, permitiéndonos captar más detalles de la distribución de los datos sin perder claridad.

---

**Paso 3: Calcular el tamaño de las clases**

El **tamaño de cada clase** se calcula dividiendo la **amplitud de variación** entre el número de clases.

#### **Fórmula**:

$
\text{Tamaño de Clase} = \frac{\text{Amplitud de Variación}}{\text{Número de Clases}}
$

#### **Ejemplo en Python**:

```python
# Calcular el tamaño de clase
tamano_clase = round(amplitud_variacion / clases_sturges, 2)
print(f"Tamaño de cada clase: {tamano_clase}")
```

Esto significa que cada clase abarcará un rango de **2.86 unidades** (por ejemplo, de 76.00 a 78.86, de 78.86 a 81.71, etc.).

#### **Paso 4: Definir los límites de las clases**

Usando el tamaño de clase calculado, podemos definir los **límites inferiores** y **superiores** de cada clase. La primera clase empieza en el valor mínimo, y las siguientes clases se definen sumando el tamaño de clase de forma sucesiva.

#### **Ejemplo en Python**:

```python
# Definir los límites inferiores y superiores de las clases
limites_inferiores = np.arange(minimo, maximo, tamano_clase)
limites_superiores = limites_inferiores + tamano_clase

# Mostrar los límites de clase
for i in range(len(limites_inferiores)):
    print(f"Clase {i+1}: {limites_inferiores[i]:.2f} - {limites_superiores[i]:.2f}")
```

#### **Paso 5: Contar la frecuencia en cada clase**

Utilizando estos intervalos de clase, contamos cuántos datos caen dentro de cada clase para obtener la **frecuencia absoluta**.

#### **Ejemplo en Python**:

```python
# Crear la tabla de frecuencias
frecuencias_absolutas, _ = np.histogram(data, bins=np.arange(minimo, maximo+tamano_clase, tamano_clase))

# Mostrar frecuencias absolutas
for i in range(len(frecuencias_absolutas)):
    print(f"Clase {i+1}: {limites_inferiores[i]:.2f} - {limites_superiores[i]:.2f}, Frecuencia absoluta: {frecuencias_absolutas[i]}")
```

#### **Paso 6: Calcular la frecuencia relativa**

La **frecuencia relativa** es la proporción de datos en cada clase con respecto al total de datos. Se calcula dividiendo la frecuencia absoluta entre el número total de datos.

#### **Fórmula**:

$
\text{Frecuencia Relativa} = \frac{\text{Frecuencia Absoluta}}{n}
$

#### **Ejemplo en Python**:

```python
# Calcular la frecuencia relativa
frecuencia_relativa = frecuencias_absolutas / n

# Mostrar frecuencia relativa
for i in range(len(frecuencia_relativa)):
    print(f"Clase {i+1}: {limites_inferiores[i]:.2f} - {limites_superiores[i]:.2f}, Frecuencia relativa: {frecuencia_relativa[i]:.2f}")
```

#### **Paso 7: Calcular la frecuencia acumulada**

La **frecuencia acumulada** es la suma de las frecuencias absolutas hasta un determinado punto, y muestra cuántos datos caen por debajo de ciertos límites de clase.

#### **Ejemplo en Python**:

```python
# Calcular la frecuencia acumulada
frecuencia_acumulada = np.cumsum(frecuencias_absolutas)

# Mostrar frecuencia acumulada
for i in range(len(frecuencia_acumulada)):
    print(f"Clase {i+1}: {limites_inferiores[i]:.2f} - {limites_superiores[i]:.2f}, Frecuencia acumulada: {frecuencia_acumulada[i]}")
```

### **Tabla de Frecuencias Final**

Podemos organizar todo en una tabla para visualizar mejor los resultados:

```python
import pandas as pd

# Crear un DataFrame para organizar la tabla de frecuencias
tabla_frecuencias = pd.DataFrame({
    'Clase': [f"{limites_inferiores[i]:.2f} - {limites_superiores[i]:.2f}" for i in range(len(limites_inferiores))],
    'Frecuencia Absoluta': frecuencias_absolutas,
    'Frecuencia Relativa': frecuencia_relativa,
    'Frecuencia Acumulada': frecuencia_acumulada
})

# Mostrar la tabla de frecuencias
tabla_frecuencias
```

Este proceso automatizado en Python nos permite construir una **tabla de frecuencias** paso a paso, utilizando la **fórmula de Sturges** para determinar el número de clases. La tabla de frecuencias final nos ofrece una descripción clara y completa de cómo están distribuidos los datos, facilitando el análisis de **distribuciones** y el cálculo de **medidas de tendencia central** o **dispersión**.

5.2 **Histograma**

#### **Teoría**:

Un **histograma** es un tipo de gráfico de barras utilizado para representar la **distribución de frecuencias** de un conjunto de datos. Los datos se agrupan en **clases** o **intervalos** (que pueden ser determinados mediante métodos como la **fórmula de Sturges**), y la altura de cada barra representa la **frecuencia absoluta** o **frecuencia relativa** de los datos dentro de ese intervalo.

### **Características principales de un histograma**:

- **Clases o intervalos**: Los datos se agrupan en intervalos contiguos, de igual tamaño, para facilitar la visualización de la distribución.
- **Altura de las barras**: La altura de cada barra es proporcional al número de datos (frecuencia) que caen dentro del intervalo correspondiente.
- **Contigüidad**: A diferencia de un gráfico de barras, en un histograma las barras están **contiguas** entre sí, lo que refleja que los datos están organizados en intervalos continuos.
- **Datos continuos**: Los histogramas son especialmente útiles para datos **continuos**, como temperaturas, tiempos o calificaciones, donde los valores pueden tomar cualquier valor dentro de un rango.

---

### **Pasos para construir un histograma**

1. **Determinar las clases o intervalos**: Como hicimos en el proceso de la **tabla de frecuencias**, determinamos el número de clases utilizando la **fórmula de Sturges** o dividiendo el rango de los datos en intervalos de igual tamaño.
2. **Contar la frecuencia en cada clase**: Una vez definidos los intervalos, contamos cuántos datos caen dentro de cada clase o intervalo.
3. **Graficar el histograma**: Se representa el **intervalo** en el eje X y la **frecuencia** (o frecuencia relativa) en el eje Y.

---

### **Ejemplo en Python: Construcción de un histograma**

Vamos a utilizar el mismo conjunto de datos de calificaciones de estudiantes que hemos usado anteriormente. Primero, determinaremos las clases utilizando la **fórmula de Sturges** y luego graficaremos el histograma.

#### **Paso 1: Determinar las clases y frecuencias**

Para determinar los intervalos o clases, utilizaremos los pasos que seguimos en la **tabla de frecuencias** (determinación de clases por Sturges, cálculo del tamaño de clase, y conteo de frecuencias).

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Conjunto de datos: calificaciones de un grupo de estudiantes
data = [85, 90, 78, 92, 88, 76, 95, 89, 85, 90, 91, 84, 93, 87, 77, 96, 85, 90, 88, 86,
        92, 81, 89, 93, 90, 79, 85, 87, 92, 94, 83, 88, 86, 95, 91, 80, 84, 82, 94, 89,
        88, 90, 92, 85, 96, 87, 85, 93, 91, 89]

# Determinamos el valor mínimo, máximo y la amplitud de variación
minimo = min(data)
maximo = max(data)
amplitud_variacion = maximo - minimo

# Número total de datos
n = len(data)

# Fórmula de Sturges para determinar el número de clases
clases_sturges = round(1 + 3.322 * math.log10(n))

# Calcular el tamaño de clase
tamano_clase = round(amplitud_variacion / clases_sturges, 2)

# Mostrar el número de clases y el tamaño de clase
print(f"Número de clases (Sturges): {clases_sturges}")
print(f"Tamaño de clase: {tamano_clase}")
```

#### **Paso 2: Graficar el histograma**

Ahora que conocemos el número de clases y el tamaño de cada clase, utilizamos la función `plt.hist()` de `matplotlib` para crear el histograma. El número de **bins** será el número de clases, y la **frecuencia** de cada clase será calculada automáticamente.

```python
# Graficar el histograma
plt.hist(data, bins=clases_sturges, edgecolor='black', color='skyblue')

# Añadir título y etiquetas
plt.title('Histograma de Calificaciones', fontsize=14)
plt.xlabel('Calificaciones', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

**Resultado del gráfico**:

Este código genera un histograma con **7 barras**, donde cada barra representa un intervalo de calificaciones (definido por el tamaño de clase de 2.86 unidades), y la altura de cada barra muestra la **frecuencia** de las calificaciones que caen dentro de ese intervalo.

### **Interpretación del Histograma**:

- El **eje X** muestra los intervalos de calificaciones (por ejemplo, de **76.00 a 78.86**, de **78.86 a 81.71**, etc.).
- El **eje Y** muestra la **frecuencia absoluta**, es decir, cuántas calificaciones caen en cada intervalo.
- El histograma nos permite observar la **distribución** de las calificaciones:
  - Si las barras más altas están a la izquierda, la distribución podría estar sesgada hacia valores más bajos.
  - Si las barras más altas están en el centro, la distribución podría ser más simétrica, con valores concentrados alrededor de la media.

---

### **Paso 3: Añadir la frecuencia relativa (opcional)**

Si queremos que el histograma muestre la **frecuencia relativa** en lugar de la frecuencia absoluta, podemos agregar el argumento `density=True` en la función `plt.hist()`. Esto normaliza las barras de modo que el área total del histograma sea igual a 1.

```python
# Graficar el histograma con frecuencias relativas
plt.hist(data, bins=clases_sturges, edgecolor='black', color='skyblue', density=True)

# Añadir título y etiquetas
plt.title('Histograma de Calificaciones (Frecuencia Relativa)', fontsize=14)
plt.xlabel('Calificaciones', fontsize=12)
plt.ylabel('Frecuencia Relativa', fontsize=12)

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

**Interpretación**:

- Ahora, el eje Y muestra la **frecuencia relativa**, es decir, la proporción de datos que caen en cada intervalo con respecto al total.
- La **suma del área de todas las barras** es igual a **1**, lo que significa que estamos trabajando con proporciones.

---

### **Conclusión**

El **histograma** es una herramienta visual poderosa que nos permite observar la **distribución de los datos** de manera rápida y efectiva. Gracias a su estructura, es ideal para trabajar con datos **continuos** o cuando se necesita visualizar datos agrupados en intervalos.

- **Ventajas**:
  
  - Nos permite identificar la **forma de la distribución**: si es **simétrica**, **sesgada** a la derecha o izquierda, o si tiene múltiples picos.
  - Es útil para detectar **valores atípicos** o **outliers**.
  - Facilita la visualización de grandes conjuntos de datos al agruparlos en intervalos.
- **Limitaciones**:
  
  - La elección del número de clases o intervalos puede afectar la interpretación del histograma. Si hay muy pocas clases, podemos perder detalle, y si hay demasiadas, el gráfico puede volverse ruidoso.
  - Para datos **discretos** (como conteos de enteros), puede no ser tan útil como otras visualizaciones (como gráficos de barras o gráficos de puntos).

El histograma es una herramienta fundamental en el análisis de datos, ya que proporciona una visión clara de cómo los datos están distribuidos y facilita la comprensión de su comportamiento.

## Gráfico de Barras

### **Definición y Características**

El **gráfico de barras** es una herramienta de visualización de datos que utiliza barras rectangulares para representar la frecuencia, cantidad o cualquier otra métrica de diferentes categorías. Cada barra en el gráfico representa una categoría distinta y su longitud o altura es proporcional al valor que representa.

### **Componentes de un Gráfico de Barras**

1. **Ejes**:
   - **Eje X (Horizontal)**: Representa las categorías o grupos que se están comparando.
   - **Eje Y (Vertical)**: Representa la métrica cuantitativa que se está midiendo (por ejemplo, frecuencia, cantidad, porcentaje).

2. **Barras**:
   - **Longitud/Altura**: Proporcional al valor que representa.
   - **Color**: Puede utilizarse para diferenciar categorías o resaltar información específica.
   - **Espaciado**: Las barras están separadas entre sí para enfatizar que representan categorías distintas.

3. **Título**:
   - Describe el propósito del gráfico.

4. **Etiquetas de los Ejes**:
   - **Eje X**: Nombre de las categorías.
   - **Eje Y**: Nombre de la métrica cuantitativa.

5. **Leyenda** (opcional):
   - Explica el significado de diferentes colores o patrones utilizados en las barras.

### **Tipos de Gráficos de Barras**

1. **Gráfico de Barras Verticales**:
   - Las barras son verticales y se utilizan comúnmente para comparar diferentes categorías.

2. **Gráfico de Barras Horizontales**:
   - Las barras son horizontales y son útiles cuando las etiquetas de las categorías son largas o hay muchas categorías.

3. **Gráfico de Barras Apiladas**:
   - Muestra la composición de cada categoría dividiendo las barras en segmentos apilados que representan subcategorías.

4. **Gráfico de Barras Agrupadas**:
   - Agrupa barras de diferentes categorías juntas para facilitar la comparación entre ellas.

### **Ventajas del Gráfico de Barras**

- **Claridad**: Facilita la comparación directa entre diferentes categorías.
- **Flexibilidad**: Puede adaptarse a diferentes tipos de datos y propósitos.
- **Visualmente Intuitivo**: Fácil de interpretar para la mayoría de las audiencias.
- **Versatilidad**: Adecuado tanto para datos categóricos como discretos.

### **Limitaciones del Gráfico de Barras**

- **Espacio**: Puede volverse desordenado o difícil de leer si hay demasiadas categorías.
- **Comparación Limitada**: Menos efectivo para mostrar cambios a lo largo del tiempo en comparación con gráficos de líneas.
- **Percepción de Área**: La longitud de las barras puede ser menos precisa para comparar diferencias pequeñas en valores.

### **Cuándo Utilizar un Gráfico de Barras**

- **Comparar Categorías**: Ideal para comparar diferentes grupos o categorías de manera directa.
- **Mostrar Cambios Relativos**: Útil para resaltar diferencias relativas entre categorías.
- **Visualizar Distribuciones Categóricas**: Perfecto para representar la frecuencia o proporción de categorías específicas.

### **Ejemplo de Código: Gráfico de Barras en Python con Matplotlib**

A continuación, se presenta un ejemplo de cómo crear un gráfico de barras utilizando la biblioteca **Matplotlib** en Python. Este ejemplo ilustra cómo representar la frecuencia de calificaciones en diferentes rangos.

### 5.3 **Ojiva**

#### **Teoría**:

La **ojiva** es un tipo de gráfico que muestra la **frecuencia acumulada** de un conjunto de datos. Representa cómo se acumulan los datos conforme avanzamos por los valores, permitiendo visualizar el número total de observaciones por debajo de un cierto valor.

A diferencia de un **histograma**, que muestra la distribución de los datos en intervalos, la ojiva se utiliza para observar cómo se **acumulan** los datos a medida que recorremos el rango de valores. Esto es especialmente útil para calcular **percentiles** y analizar cómo se distribuyen los datos en diferentes partes del conjunto.

### **Características principales de la Ojiva**:

- La ojiva es un gráfico de **línea** que muestra la **frecuencia acumulada** en el eje Y y los **valores de los datos** en el eje X.
- Es útil para analizar la **distribución acumulativa** de los datos y para encontrar valores como el **percentil 50** (mediana), el **percentil 25** o el **percentil 90**.
- Nos permite responder preguntas como: "¿Cuántos datos son menores que un valor determinado?" o "¿En qué rango se encuentran los primeros 50% de los datos?".

---

### **Pasos para construir una Ojiva**

1. **Determinar las clases** o **intervalos** en los que se agruparán los datos (igual que en la tabla de frecuencias).
2. **Calcular la frecuencia acumulada** de los datos en cada intervalo.
3. **Graficar la ojiva**, colocando los puntos de cada intervalo y su correspondiente frecuencia acumulada en el gráfico.

---

### **Ejemplo en Python: Construcción de una Ojiva**

Para construir la ojiva, utilizaremos el mismo conjunto de **calificaciones de estudiantes** que hemos venido usando, y seguiremos el mismo proceso para calcular las clases y frecuencias, pero ahora nos enfocaremos en la **frecuencia acumulada**.

#### **Paso 1: Determinar las clases y la frecuencia acumulada**

Utilizaremos la fórmula de Sturges para determinar las clases y luego calcularemos la frecuencia acumulada.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Conjunto de datos: calificaciones de un grupo de estudiantes
data = [85, 90, 78, 92, 88, 76, 95, 89, 85, 90, 91, 84, 93, 87, 77, 96, 85, 90, 88, 86,
        92, 81, 89, 93, 90, 79, 85, 87, 92, 94, 83, 88, 86, 95, 91, 80, 84, 82, 94, 89,
        88, 90, 92, 85, 96, 87, 85, 93, 91, 89]

# Determinar el valor mínimo, máximo y la amplitud de variación
minimo = min(data)
maximo = max(data)
amplitud_variacion = maximo - minimo

# Número total de datos
n = len(data)

# Fórmula de Sturges para determinar el número de clases
clases_sturges = round(1 + 3.322 * math.log10(n))

# Calcular el tamaño de clase
tamano_clase = round(amplitud_variacion / clases_sturges, 2)

# Crear los intervalos de clase
limites_inferiores = np.arange(minimo, maximo, tamano_clase)
limites_superiores = limites_inferiores + tamano_clase

# Crear la tabla de frecuencias
frecuencias_absolutas, _ = np.histogram(data, bins=np.arange(minimo, maximo+tamano_clase, tamano_clase))

# Calcular la frecuencia acumulada
frecuencia_acumulada = np.cumsum(frecuencias_absolutas)

# Mostrar las clases y la frecuencia acumulada
for i in range(len(frecuencia_acumulada)):
    print(f"Clase {i+1}: {limites_inferiores[i]:.2f} - {limites_superiores[i]:.2f}, Frecuencia acumulada: {frecuencia_acumulada[i]}")
```

En este resultado, vemos cómo la **frecuencia acumulada** nos muestra cuántos datos se han contado hasta cada clase. Por ejemplo, **3** estudiantes tienen una calificación entre **76.00 y 78.86**, y un total de **50** estudiantes han sido contabilizados hasta el intervalo más alto.

---

#### **Paso 2: Graficar la Ojiva**

Para graficar la ojiva, simplemente colocamos en el eje **X** los límites superiores de las clases y en el eje **Y** la **frecuencia acumulada** correspondiente.

```python
# Graficar la Ojiva (Frecuencia Acumulada)
plt.plot(limites_superiores, frecuencia_acumulada, marker='o', linestyle='-', color='blue')

# Añadir título y etiquetas
plt.title('Ojiva de Frecuencia Acumulada', fontsize=14)
plt.xlabel('Calificaciones', fontsize=12)
plt.ylabel('Frecuencia Acumulada', fontsize=12)

# Mostrar la cuadrícula y el gráfico
plt.grid(True)
plt.show()
```

Interpretación del gráfico**:

- La **ojiva** comienza en el primer valor de la clase inferior y continúa hacia la derecha, mostrando cómo se acumulan las frecuencias a medida que se avanzan por los intervalos de datos.
- El eje **X** representa las **calificaciones**, y el eje **Y** muestra cuántos estudiantes tienen calificaciones menores o iguales al valor de la clase superior en el eje X.
- Este gráfico es útil para identificar valores como el **percentil 50** (la mediana), que corresponde al valor en el que la frecuencia acumulada alcanza el 50% del total.

---

### **Paso 3: Identificar Percentiles en la Ojiva (Opcional)**

Una de las grandes ventajas de la ojiva es que nos permite **calcular percentiles**, como el **percentil 25**, el **percentil 50** (mediana) o el **percentil 75**.

Por ejemplo, el **percentil 50** es el punto donde la frecuencia acumulada es igual al 50% del total de datos.

#### **Cálculo del Percentil 50 (Mediana)**:

```python
# Calcular la mediana (percentil 50)
mediana = np.percentile(data, 50)
print(f"Mediana (Percentil 50): {mediana}")
```

La **ojiva** es una herramienta visual importante para **analizar la acumulación de datos** y observar cómo se distribuyen a lo largo de un conjunto. Nos permite ver cómo se suman las frecuencias en cada clase o intervalo y facilita la identificación de **percentiles** y valores acumulativos importantes.

- **Ventajas de la Ojiva**:
  
  - Permite identificar fácilmente los **percentiles** de la distribución.
  - Visualiza cómo se **acumulan los datos** a medida que avanzamos por los valores.
  - Es útil para comparar con otros tipos de gráficos, como el **histograma**, para obtener una visión más completa de la distribución.
- **Limitaciones**:
  
  - La ojiva no muestra la **distribución detallada** dentro de los intervalos como lo hace un histograma, por lo que no es ideal si queremos ver cómo se distribuyen los datos dentro de cada clase.

Al igual que el histograma, la **ojiva** es una herramienta fundamental para el análisis exploratorio de datos y es especialmente útil cuando se quiere entender cómo los datos se acumulan a lo largo de un conjunto.

### 5.4 **Indicadores de Forma: Asimetría y Curtosis**

#### **Teoría**

Los **indicadores de forma** nos permiten conocer cómo es la **distribución** de nuestros datos en términos de su **asimetría** y **curtosis**. Estas métricas son esenciales para entender si una distribución se desvía de la simetría o tiene características extremas como colas largas o picos muy marcados. Entender estos indicadores nos ayuda a interpretar mejor los datos y detectar comportamientos atípicos.

---

#### **Asimetría (Skewness)**

La **asimetría** mide la **simetría** de una distribución:

- **Asimetría positiva**: La distribución tiene una **cola más larga a la derecha**. Esto significa que hay algunos valores inusualmente altos que hacen que la cola derecha se alargue.
- **Asimetría negativa**: La distribución tiene una **cola más larga a la izquierda**, lo que indica que hay valores inusualmente bajos.
- **Asimetría cero (0)**: La distribución es **perfectamente simétrica**.

#### **Curtosis**

La **curtosis** mide la **forma del pico** de una distribución:

- **Curtosis positiva**: La distribución tiene un pico más alto y colas más gruesas. Esto significa que los datos están muy concentrados alrededor de la media y hay más probabilidades de valores extremos en las colas.
- **Curtosis negativa**: La distribución es más **plana**, con menos datos en el centro y colas más delgadas, lo que significa que hay menos valores extremos.

---

### **Intuición Visual y Ejemplificación**

Para entender mejor estos conceptos, vamos a comparar distribuciones con **asimetría** y **curtosis** en diferentes escenarios. Primero calcularemos las métricas y luego visualizaremos las distribuciones.

---

### **Ejemplo en Python: Asimetría y Curtosis**

A continuación, vamos a generar distribuciones que exhiben diferentes tipos de asimetría y curtosis para visualizar cómo estos indicadores afectan las formas de las distribuciones.

#### **Paso 1: Generar ejemplos de distribuciones**

Vamos a crear diferentes conjuntos de datos que tengan:

- **Distribución normal (simétrica)**
- **Distribución con asimetría positiva**
- **Distribución con asimetría negativa**
- **Distribución con curtosis positiva (picos altos)**
- **Distribución con curtosis negativa (picos bajos)**

#### **Código**:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Generar distribuciones
np.random.seed(42)
data_normal = np.random.normal(0, 1, 1000)  # Distribución normal
data_skew_pos = np.random.exponential(scale=2, size=1000)  # Asimetría positiva
data_skew_neg = -np.random.exponential(scale=2, size=1000)  # Asimetría negativa
data_kurt_high = np.random.laplace(loc=0, scale=0.5, size=1000)  # Curtosis positiva (pico alto)
data_kurt_low = np.random.uniform(-3, 3, 1000)  # Curtosis negativa (pico bajo)

# Calcular asimetría y curtosis
skew_normal_value = skew(data_normal)
kurt_normal_value = kurtosis(data_normal)
skew_pos_value = skew(data_skew_pos)
kurt_pos_value = kurtosis(data_skew_pos)
skew_neg_value = skew(data_skew_neg)
kurt_neg_value = kurtosis(data_skew_neg)
kurt_high_value = kurtosis(data_kurt_high)
kurt_low_value = kurtosis(data_kurt_low)

# Crear la figura y los subgráficos en una sola columna
fig, axs = plt.subplots(5, 1, figsize=(8, 30))  # Aumentar más el tamaño de la figura para más espacio

# Configurar espaciado entre gráficos
plt.subplots_adjust(hspace=0.5)

# Distribución normal
axs[0].hist(data_normal, bins=30, color='lightblue', edgecolor='black')
axs[0].set_title(f"Distribución Normal\nAsimetría: {skew_normal_value:.2f}, Curtosis: {kurt_normal_value:.2f}")
axs[0].annotate("Interpretación: La distribución es simétrica, con un pico moderado y colas normales.\n"
                "La asimetría y curtosis son cercanas a 0, lo que indica una distribución normal estándar.",
                xy=(0.5, -0.25), xycoords='axes fraction', ha='center', fontsize=10)

# Asimetría positiva
axs[1].hist(data_skew_pos, bins=30, color='lightcoral', edgecolor='black')
axs[1].set_title(f"Asimetría Positiva\nAsimetría: {skew_pos_value:.2f}, Curtosis: {kurt_pos_value:.2f}")
axs[1].annotate("Interpretación: La distribución tiene una cola larga a la derecha, lo que indica valores altos extremos.\n"
                "La curtosis alta sugiere un pico pronunciado y colas gruesas.",
                xy=(0.5, -0.25), xycoords='axes fraction', ha='center', fontsize=10)

# Asimetría negativa
axs[2].hist(data_skew_neg, bins=30, color='lightgreen', edgecolor='black')
axs[2].set_title(f"Asimetría Negativa\nAsimetría: {skew_neg_value:.2f}, Curtosis: {kurt_neg_value:.2f}")
axs[2].annotate("Interpretación: La distribución tiene una cola larga a la izquierda, con valores bajos extremos.\n"
                "El pico es elevado, y las colas son más gruesas de lo habitual.",
                xy=(0.5, -0.25), xycoords='axes fraction', ha='center', fontsize=10)

# Curtosis alta (pico alto)
axs[3].hist(data_kurt_high, bins=30, color='deepskyblue', edgecolor='black')
axs[3].set_title(f"Curtosis Alta (Pico Alto)\nAsimetría: {skew(data_kurt_high):.2f}, Curtosis: {kurt_high_value:.2f}")
axs[3].annotate("Interpretación: La distribución es simétrica pero tiene un pico muy alto y colas gruesas.\n"
                "Los datos están muy concentrados alrededor de la media, pero hay valores extremos.",
                xy=(0.5, -0.25), xycoords='axes fraction', ha='center', fontsize=10)

# Curtosis baja (pico bajo)
axs[4].hist(data_kurt_low, bins=30, color='purple', edgecolor='black')
axs[4].set_title(f"Curtosis Baja (Pico Bajo)\nAsimetría: {skew(data_kurt_low):.2f}, Curtosis: {kurt_low_value:.2f}")
axs[4].annotate("Interpretación: La distribución es más plana y los datos están dispersos de manera uniforme.\n"
                "Las colas son delgadas, lo que sugiere menos valores extremos.",
                xy=(0.5, -0.25), xycoords='axes fraction', ha='center', fontsize=10)

# Mostrar la gráfica
plt.show()
```

**Casos Extremos y la Intuición detrás de Asimetría y Curtosis**

#### **Asimetría**:

1. **Positiva**: Imagine una situación en la que la mayoría de los estudiantes tienen calificaciones cercanas a 70, pero algunos pocos tienen calificaciones extremadamente altas (95 o más). Esto crea una cola hacia la derecha.
2. **Negativa**: Si la mayoría de los estudiantes tienen calificaciones altas, pero algunos pocos tienen calificaciones muy bajas, la distribución tendrá una cola a la izquierda.

#### **Curtosis**:

1. **Alta Curtosis**: Se asocia con la probabilidad de observar valores extremos más alta que en una distribución normal. Esto es útil en situaciones como mercados financieros, donde las crisis pueden generar picos agudos de valores extremos (por ejemplo, grandes caídas o subidas en los precios).
2. **Baja Curtosis**: Si los datos están más dispersos y no tienen picos extremos, como por ejemplo, en el tamaño de los zapatos de una población general, la curtosis será baja.

### **Interpretación Final**

- **Asimetría** nos ayuda a entender si nuestros datos están más inclinados hacia valores bajos o altos.
- **Curtosis** nos indica si los datos están concentrados alrededor de la media o si tienden a tener más valores extremos.

Estos **indicadores de forma** son esenciales para interpretar la **distribución** de un conjunto de datos, lo que puede ser clave en muchos análisis estadísticos, especialmente cuando se espera que los datos se distribuyan de manera **normal** pero se observa que tienen **asimetría** o **colas más gruesas** de lo esperado.

# 6. **Diagrama de Cajas y Brazos**

#### **Teoría**:

El **diagrama de cajas y brazos** (o **boxplot**) es una herramienta gráfica que permite representar de manera compacta la **distribución** de un conjunto de datos, resaltando puntos clave como la **mediana**, los **cuartiles** y los **valores atípicos**. Este diagrama es muy útil para identificar la **dispersión**, **asimetría** y la **presencia de outliers** (valores atípicos).

#### **Elementos Clave del Diagrama de Cajas**:

1. **Caja**: La caja representa el **rango intercuartílico (IQR)**, es decir, el rango entre el primer cuartil (**Q1**, el percentil 25) y el tercer cuartil (**Q3**, el percentil 75). Este rango contiene el **50% central** de los datos.
  
2. **Línea en la caja**: Representa la **mediana** (el percentil 50). Es el valor que divide la mitad inferior de la mitad superior de los datos.
  
3. **Brazos (o "bigotes")**: Los brazos del diagrama se extienden hasta los valores que están dentro de 1.5 veces el IQR desde los cuartiles (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR). Todo valor fuera de este rango se considera **atípico**.
  
4. **Puntos atípicos**: Estos son los valores que caen fuera del rango de los brazos y son indicados por puntos individuales fuera de los "bigotes". Indican **outliers** o **valores atípicos**.
  

#### **Ejemplo de Código en Python**:

A continuación se presenta un ejemplo para visualizar un diagrama de cajas utilizando **Python**:

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo (calificaciones de un grupo de estudiantes)
data = [85, 90, 78, 92, 88, 76, 95, 89, 85, 100, 45]  # Incluyendo un valor atípico (45)

# Crear la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Crear el diagrama de cajas
ax.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))  # Gráfico de cajas horizontal

# Añadir los puntos individuales debajo del diagrama de cajas (distribución unidimensional)
y = np.ones(len(data))  # Todos los puntos en la misma fila (en el eje y)


# Ajustes adicionales
ax.set_yticks([])  # Ocultar el eje y ya que solo necesitamos el eje x
ax.set_xlabel('Calificaciones')
ax.set_title('Diagrama de Cajas y Distribución de Datos (Dot Plot)', fontsize=14)
ax.grid(True, axis='x')  # Mostrar una cuadrícula en el eje x

# Mostrar la gráfica
plt.show()
```

3. **Brazos**: Se extienden hasta los puntos que se encuentran dentro de 1.5 veces el rango intercuartílico (IQR).
4. **Valores atípicos**: En este caso, el valor **45** sería un outlier, ya que cae fuera del rango esperado de los datos.

#### **Interpretación del Diagrama**:

- Si la **caja** es muy alargada, significa que hay mucha **dispersión** en los datos.

- La **longitud de los brazos** indica el **rango** de los datos. Brazos muy largos sugieren datos muy dispersos, mientras que brazos cortos indican una menor dispersión.
- **Outliers**: Los valores atípicos se identifican como puntos individuales fuera del rango de los brazos y pueden dar información importante sobre posibles errores o comportamientos atípicos en los datos.

- Comparar distribuciones de varios conjuntos de datos simultáneamente.

La **tabla de contingencias** es una herramienta esencial para analizar la relación entre dos variables categóricas. Permite organizar los datos en una matriz donde cada celda representa la frecuencia de una combinación específica de categorías de las dos variables.

##### **Ejemplo en Python: Encuesta sobre Educación y Propiedad de Automóvil**

Supongamos que realizamos una encuesta a 50 personas sobre su nivel de educación y si poseen un automóvil.

```python
import pandas as pd

# Datos de ejemplo: nivel educativo y si tienen automóvil
data = {
    'Educación': [
        'Secundaria', 'Preparatoria', 'Licenciatura', 'Licenciatura', 'Secundaria',
        'Preparatoria', 'Preparatoria', 'Secundaria', 'Licenciatura', 'Doctorado',
        'Doctorado', 'Secundaria', 'Preparatoria', 'Licenciatura', 'Doctorado',
        'Secundaria', 'Preparatoria', 'Licenciatura', 'Doctorado', 'Licenciatura',
        'Preparatoria', 'Secundaria', 'Licenciatura', 'Doctorado', 'Secundaria',
        'Preparatoria', 'Licenciatura', 'Doctorado', 'Secundaria', 'Preparatoria',
        'Licenciatura', 'Doctorado', 'Secundaria', 'Preparatoria', 'Licenciatura',
        'Doctorado', 'Secundaria', 'Preparatoria', 'Licenciatura', 'Doctorado',
        'Secundaria', 'Preparatoria', 'Licenciatura', 'Doctorado', 'Secundaria',
        'Preparatoria', 'Licenciatura', 'Doctorado', 'Secundaria', 'Secundaria'
    ],
    'Tiene_Automóvil': [
        'Sí', 'No', 'Sí', 'No', 'Sí',
        'Sí', 'No', 'No', 'Sí', 'Sí',
        'No', 'Sí', 'No', 'Sí', 'No',
        'Sí', 'No', 'Sí', 'Sí', 'No',
        'No', 'Sí', 'Sí', 'No', 'Sí',
        'No', 'Sí', 'Sí', 'No', 'Sí',
        'No', 'Sí', 'Sí', 'No', 'Sí',
        'No', 'Sí', 'Sí', 'No', 'Sí',
        'No', 'Sí', 'Sí', 'No', 'Sí',
        'No', 'Sí', 'Sí', 'No', 'Sí'
    ]
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Crear la tabla de contingencias
tabla_contingencia = pd.crosstab(df['Educación'], df['Tiene_Automóvil'])

# Mostrar la tabla
print(tabla_contingencia)
```

##### **Interpretación**:

- **Doctorado**: 5 personas no tienen automóvil y 6 sí lo tienen.
- **Licenciatura**: 4 personas no tienen automóvil y 9 sí lo tienen.
- **Preparatoria**: 8 personas no tienen automóvil y 4 sí lo tienen.
- **Secundaria**: 4 personas no tienen automóvil y 10 sí lo tienen.

Esta tabla nos permite observar cómo se distribuye la propiedad de automóviles según el nivel educativo de las personas encuestadas.

##### **Visualización Adicional: Mapa de Calor de la Tabla de Contingencias**

Para una mejor visualización, podemos representar la tabla de contingencias como un **mapa de calor**.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
plt.figure(figsize=(8, 6))

# Crear el mapa de calor
sns.heatmap(tabla_contingencia, annot=True, fmt='d', cmap='YlGnBu')

# Títulos y etiquetas
plt.title('Mapa de Calor: Educación vs. Propiedad de Automóvil')
plt.xlabel('Tiene Automóvil')
plt.ylabel('Nivel Educativo')

# Mostrar el gráfico
plt.show()
```

**Interpretación**:
El mapa de calor utiliza colores para representar la intensidad de las frecuencias. Las celdas más oscuras indican una mayor frecuencia, lo que facilita la identificación de patrones en los datos.

#### 2. **Frecuencias Conjuntas y Marginales**

- **Frecuencias Conjuntas**: Son las frecuencias que se encuentran en el interior de la tabla de contingencias, representando la cantidad de observaciones para cada combinación de categorías de las dos variables.
  
- **Frecuencias Marginales**: Son los totales por fila y por columna, que muestran las frecuencias de cada categoría de una variable independientemente de la otra.
  

##### **Continuando con el Ejemplo Anterior:**

```python
# Crear tabla de contingencias con márgenes
tabla_con_marginales = pd.crosstab(df['Educación'], df['Tiene_Automóvil'], margins=True)

# Mostrar la tabla con frecuencias marginales
print(tabla_con_marginales)
```

**Interpretación**:

- **Frecuencias Conjuntas**: Se encuentran en el cuerpo de la tabla (por ejemplo, 10 personas con Licenciatura tienen automóvil).
- **Frecuencias Marginales**:
  - **Filas (Educación)**:
    - **Doctorado**: Total de 11 personas (5 No, 6 Sí).
    - **Licenciatura**: Total de 13 personas (4 No, 9 Sí).
    - **Preparatoria**: Total de 12 personas (8 No, 4 Sí).
    - **Secundaria**: Total de 14 personas (4 No, 10 Sí).
  - **Columnas (Tiene Automóvil)**:
    - **No**: Total de 21 personas.
    - **Sí**: Total de 29 personas.

Estas frecuencias marginales nos permiten entender la distribución general de cada variable sin considerar la otra.

##### **Gráfico de Barras Apiladas para Frecuencias Conjuntas**

```python
# Crear el gráfico de barras apiladas
tabla_contingencia.plot(kind='bar', stacked=True, figsize=(10,7), color=['salmon', 'skyblue'])

# Títulos y etiquetas
plt.title('Propiedad de Automóvil por Nivel Educativo')
plt.xlabel('Nivel Educativo')
plt.ylabel('Número de Personas')
plt.legend(title='Tiene Automóvil')

# Mostrar el gráfico
plt.show()
```

**Interpretación**:
Este gráfico muestra la distribución de la propiedad de automóviles dentro de cada nivel educativo, permitiendo comparar visualmente las proporciones entre las categorías.

#### 3. **Diagrama de Dispersión**

El **diagrama de dispersión** es una herramienta gráfica que muestra la relación entre dos variables numéricas. Cada punto en el gráfico representa un par de valores (X, Y), lo que permite visualizar patrones, tendencias o correlaciones entre ellas.

##### **Ejemplo en Python: Horas de Estudio y Calificaciones**

Imaginemos que tenemos datos sobre las horas que un grupo de estudiantes estudia por semana y sus calificaciones finales.

```python
import matplotlib.pyplot as plt
import numpy as np
# Datos de ejemplo: horas de estudio y calificación
horas_estudio = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
calificaciones = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]

# Crear el diagrama de dispersión
plt.figure(figsize=(8,6))
plt.scatter(horas_estudio, calificaciones, color='green', edgecolor='black', s=100)

# Títulos y etiquetas
plt.title('Diagrama de Dispersión: Horas de Estudio vs. Calificación')
plt.xlabel('Horas de Estudio por Semana')
plt.ylabel('Calificación Final')

# Añadir una línea de tendencia
m, b = np.polyfit(horas_estudio, calificaciones, 1)
plt.plot(horas_estudio, m*np.array(horas_estudio) + b, color='red')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

**Interpretación**:
El diagrama de dispersión muestra una relación positiva clara entre las horas de estudio y las calificaciones finales. La línea de tendencia roja resalta esta correlación positiva.

##### **Gráfico Interactivo con Plotly**

Para una experiencia más dinámica, podemos utilizar la librería **Plotly** para crear gráficos interactivos.

```python
import plotly.express as px

# Crear un DataFrame con los datos
df_estudio = pd.DataFrame({
    'Horas de Estudio': horas_estudio,
    'Calificación Final': calificaciones
})

# Crear el diagrama de dispersión interactivo
fig = px.scatter(df_estudio, x='Horas de Estudio', y='Calificación Final',
                 title='Diagrama de Dispersión Interactivo: Horas de Estudio vs. Calificación',
                 trendline='ols',  # Añadir línea de regresión
                 labels={'Horas de Estudio': 'Horas de Estudio por Semana',
                         'Calificación Final': 'Calificación Final'},
                 template='plotly_white')

# Mostrar el gráfico
fig.show()
```

#### 4. **Coeficiente de Correlación**

El **coeficiente de correlación** cuantifica la fuerza y la dirección de la relación lineal entre dos variables numéricas. El coeficiente de correlación de **Pearson** es el más comúnmente utilizado y su valor oscila entre -1 y 1.

- **+1**: Correlación positiva perfecta.
- **-1**: Correlación negativa perfecta.
- **0**: No hay correlación lineal.

##### **Cálculo del Coeficiente de Correlación de Pearson en Python**

Usaremos los mismos datos de horas de estudio y calificaciones.

```python
import numpy as np

# Calcular el coeficiente de correlación de Pearson
correlacion = np.corrcoef(horas_estudio, calificaciones)[0, 1]

# Mostrar el resultado
print(f"Coeficiente de correlación de Pearson: {correlacion:.2f}")
```

##### **Interpretación**:

Un coeficiente de **1.00** indica una **correlación positiva perfecta**, lo que sugiere que, en este conjunto de datos, a medida que aumentan las horas de estudio, las calificaciones también aumentan de manera perfectamente lineal.

##### **Cálculo Usando Pandas**

También podemos calcular la correlación utilizando la función `corr()` de Pandas para facilitar la lectura, especialmente cuando se manejan múltiples variables.

```python
# Crear el DataFrame
df_estudio = pd.DataFrame({
    'Horas de Estudio': horas_estudio,
    'Calificación Final': calificaciones
})

# Calcular la matriz de correlación
matriz_correlacion = df_estudio.corr()

# Mostrar la matriz de correlación
print(matriz_correlacion)
```

##### **Interpretación Adicional: Correlación en Datos Reales**

En escenarios más realistas, los datos no suelen mostrar una correlación perfecta. Consideremos otro ejemplo con datos más variados.

```python
# Datos de ejemplo: horas de estudio y calificación con más variabilidad
horas_estudio_var = [5, 7, 10, 12, 15, 18, 20, 22, 25, 28]
calificaciones_var = [55, 58, 60, 62, 65, 70, 75, 78, 80, 85]

# Calcular el coeficiente de correlación de Pearson
correlacion_var = np.corrcoef(horas_estudio_var, calificaciones_var)[0, 1]

# Mostrar el resultado
print(f"Coeficiente de correlación de Pearson: {correlacion_var:.2f}")
```

**Interpretación**:
Un coeficiente de **0.99** aún indica una **correlación positiva muy fuerte**, pero no perfecta. Esto refleja una relación estrecha pero con cierta variabilidad en los datos.
