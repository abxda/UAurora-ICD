#1. **Medidas de tendencia central**

Las **medidas de tendencia central** son herramientas estadísticas fundamentales que nos permiten describir un conjunto de datos mediante un valor representativo. Existen tres medidas principales de tendencia central: la **media**, la **mediana** y la **moda**. Cada una de ellas captura una característica diferente de la distribución de los datos, y su utilidad depende de la naturaleza de los datos y del contexto de análisis.

---

#### **1.1 Media Aritmética**

- **Definición**: La **media** (o promedio) es el valor obtenido al sumar todos los datos y dividir el resultado por la cantidad total de datos. Representa el valor típico de una distribución cuando todos los datos contribuyen por igual.
  
  $
  \text{Media} = \frac{\sum_{i=1}^{n} x_i}{n}
  $
  
  Donde $x_i$ representa cada uno de los valores y $n$ es la cantidad de datos.
  
- **Aplicación**:
  
  - La media es útil cuando los datos son distribuidos de manera uniforme o no presentan valores atípicos (outliers) que distorsionen el promedio.
  - Ejemplo: Si queremos calcular el salario promedio de un grupo de empleados o la calificación promedio de una clase, la media sería la medida más apropiada.
- **Ventajas**:
  
  - Utiliza todos los valores del conjunto de datos.
  - Fácil de calcular y comprender.
- **Limitaciones**:
  
  - Es sensible a valores atípicos (outliers), que pueden distorsionar el valor de la media y no representar adecuadamente el conjunto de datos.

**Cálculo de la Media Aritmética Paso a Paso**

**Definición Matemática**:
La **media aritmética** es el valor promedio de un conjunto de datos. Matemáticamente, la media de un conjunto de datos $ x_1, x_2, \dots, x_n $ se define como:

$
\text{Media} = \frac{1}{n} \sum_{i=1}^{n} x_i = \frac{x_1 + x_2 + x_3 + \dots + x_n}{n}
$

Donde:

- $ n $ es la cantidad total de elementos en el conjunto de datos.
- $ x_i $ representa cada uno de los valores de los datos, donde $ i $ es el índice que identifica a cada valor.
- La suma $ \sum_{i=1}^{n} x_i $ representa la adición de todos los valores en el conjunto de datos.

---

### **Explicación Detallada del Código**

1. **Definir los datos**:
  
  ```python
  data = [85, 90, 78, 92, 88, 76, 95, 89]
  ```
  
  Aquí tenemos un conjunto de datos llamado `data`, que contiene las **calificaciones** de un grupo de estudiantes. Cada número en esta lista es un valor que corresponde a la calificación de un estudiante.
  
  Si expresamos los datos usando la notación matemática, estos serían:
  $
  x_1 = 85, \quad x_2 = 90, \quad x_3 = 78, \quad x_4 = 92, \quad x_5 = 88, \quad x_6 = 76, \quad x_7 = 95, \quad x_8 = 89
  $
  Y el total de valores es \( n = 8 \), es decir, hay 8 estudiantes cuyas calificaciones estamos analizando.
  
2. **Calcular la media**:
  
  ```python
  media = np.mean(data)
  ```
  
  La función `np.mean(data)` toma todos los valores en la lista `data` y calcula la media aritmética siguiendo la fórmula descrita antes. El proceso paso a paso sería:
  
  - Sumar todos los valores de las calificaciones:
    $
    \sum_{i=1}^{n} x_i = 85 + 90 + 78 + 92 + 88 + 76 + 95 + 89 = 693
    $
    
  - Dividir la suma entre el número total de calificaciones $ n = 8 $:
    
    $
    \text{Media} = \frac{693}{8} = 86.625
    $
    
  
  Por lo tanto, la media de las calificaciones es $ 86.625 $. Esto significa que el "promedio" de las calificaciones del grupo de estudiantes es de aproximadamente 86.63 (si redondeamos a dos decimales).
  
3. **Mostrar el resultado**:
  
  ```python
  print(f"La media de las calificaciones es: {media:.2f}")
  ```
  
  El código `print(f"La media de las calificaciones es: {media:.2f}")` imprime el resultado de la media en la consola, mostrando el valor con dos decimales. El formato `:.2f` asegura que el número se presente con dos lugares después del punto decimal.

````python
import numpy as np
# Datos: Calificaciones de un grupo de estudiantes
data = [85, 90, 78, 92, 88, 76, 95, 89]
# Cálculo de la media
media = np.mean(data)
print(f"La media de las calificaciones es: {media:.2f}")
````

**Interpretación**:

En este caso, la **media** nos indica que, si distribuimos las calificaciones de manera uniforme, cada estudiante tendría una calificación de 86.63. Este valor es útil para obtener una idea general de cómo se comporta el grupo en su conjunto. Sin embargo, es importante recordar que la media puede verse afectada por valores extremos (calificaciones muy altas o muy bajas).

---

Para comprender mejor la distribución de las calificaciones, podemos crear un gráfico que muestre las calificaciones individuales y la media.

````python
import matplotlib.pyplot as plt
import numpy as np

# Calificaciones de los estudiantes
data = [85, 90, 78, 92, 88, 76, 95, 89]

# Calcular la media
media = np.mean(data)

# Visualización
plt.figure(figsize=(10,6))
plt.bar(range(1, len(data) + 1), data, color='lightblue', edgecolor='black')
plt.axhline(y=media, color='r', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')
plt.xlabel('Estudiantes')
plt.ylabel('Calificaciones')
plt.title('Calificaciones de los estudiantes y la media')
plt.legend()
plt.grid(True)
plt.show()
````

Este cálculo de la media aritmética es fundamental para entender el comportamiento general de los datos, y es una de las medidas más comunes en estadística descriptiva. Si bien proporciona información valiosa, siempre debe considerarse en conjunto con otras medidas (como la mediana o la moda) para obtener una imagen completa de los datos, especialmente en situaciones donde puede haber valores extremos que influyan en la media.

Vamos a abordar la **debilidad** de la media frente a valores atípicos o extremos en el conjunto de datos. Un valor extremo puede influir desproporcionadamente en el cálculo de la media, distorsionando su representatividad. **1.2 Mediana**

- **Definición**: La **mediana** es el valor que se encuentra en el centro de un conjunto de datos ordenado. Si el número de datos es impar, la mediana es el valor central; si es par, la mediana es el promedio de los dos valores centrales. A diferencia de la media, la mediana no se ve afectada por valores atípicos y es una mejor medida de tendencia central en distribuciones asimétricas.
  
  Matemáticamente, la mediana se define como:
  
  1. Ordenar el conjunto de datos de menor a mayor.
  2. Si $ n $ (el número de datos) es impar, la mediana es el valor en la posición $ \frac{n+1}{2} $.
  3. Si $ n $ es par, la mediana es el promedio de los valores en las posiciones $ \frac{n}{2} $ y $ \frac{n}{2} + 1 $.
  
  Por ejemplo, si \( n \) es 7:
  
  $ \text{Mediana} = x_{\frac{n+1}{2}} = x_{\frac{7+1}{2}} = x_4 $
  
  Y si \( n \) es 8:
  
  $ \text{Mediana} = \frac{x_{\frac{n}{2}} + x_{\frac{n}{2}+1}}{2} = \frac{x_4 + x_5}{2} $
  
- **Aplicación**:
  
  - La mediana es útil cuando los datos tienen **valores atípicos** o están **distribuidos de manera asimétrica**. Por ejemplo, para calcular el valor típico de los ingresos en una población donde unos pocos individuos ganan mucho más que el resto, la mediana es una mejor representación del ingreso "típico" que la media.
    
  - Ejemplo: Si queremos calcular la calificación mediana en una clase donde un estudiante ha tenido una calificación extraordinariamente baja o alta, la mediana sería más representativa del rendimiento general.
    
- **Ventajas**:
  
  - No es afectada por valores atípicos o extremos.
  - Representa mejor el "centro" en distribuciones asimétricas.
- **Limitaciones**:
  
  - No utiliza toda la información del conjunto de datos, solo los valores centrales.
  - Puede no ser tan intuitiva como la media, especialmente cuando los datos son muy dispersos.

---

1. **Definir los datos**:
  
  ```python
  data = [85, 90, 78, 92, 88, 76, 95, 89]
  ```

**1.3 Moda**

- **Definición**: La **moda** es el valor o los valores que aparecen con mayor frecuencia en un conjunto de datos. Un conjunto de datos puede tener:
  
  - Una única moda (**unimodal**), si solo un valor se repite más que los demás.
  - Dos modas (**bimodal**), si hay dos valores con la misma frecuencia más alta.
  - Más de dos modas (**multimodal**), si varios valores se repiten con la misma frecuencia máxima.
  - Si ningún valor se repite, se dice que el conjunto de datos no tiene moda.
  
  A diferencia de la media y la mediana, la moda puede aplicarse tanto a variables numéricas como categóricas, lo que la hace útil para datos cualitativos.
  
- **Aplicación**:
  
  - La moda es especialmente útil para identificar el valor más frecuente en conjuntos de datos categóricos o discretos, como encuestas, donde es importante saber la opción más elegida.
  - Ejemplo: Si estamos analizando las preferencias de colores en una encuesta, la moda nos indicaría el color que fue seleccionado con más frecuencia.
- **Ventajas**:
  
  - Es la **única medida** de tendencia central aplicable a datos categóricos (como colores, categorías, o elecciones).
  - **Fácil de calcular** e interpretar en conjuntos de datos pequeños o categóricos.
- **Limitaciones**:
  
  - Puede no ser representativa en distribuciones **uniformes** (donde no hay un valor que se repita más) o en distribuciones **multimodales** (con más de una moda).
  - No siempre refleja el centro de la distribución, especialmente en distribuciones numéricas con valores únicos.

---

### **Cálculo de la Moda Paso a Paso**

1. **Definir los datos**:
  
  ```python
  data = [85, 90, 78, 92, 88, 76, 95, 89, 85]
  ```
  
  En este conjunto de datos tenemos una repetición del valor 85, que aparece dos veces.
  
2. **Calcular la moda**:
  
  Utilizamos `scipy.stats.mode` para calcular la moda:
  
  ```python
  from scipy import stats
  
  moda = stats.mode(data)
  print(f"La moda es: {moda.mode} y aparece {moda.count} veces")
  ```
  
  Este código devuelve que la moda es **85** y que aparece **2 veces**.
  
3. **Interpretación**:
  
  La moda nos indica que la calificación más frecuente entre los estudiantes es **85**, lo que significa que **dos estudiantes** obtuvieron esta calificación.
  

---

````python
data = [85, 90, 78, 92, 88, 76, 95, 89, 85]
from scipy import stats

moda = stats.mode(data)
print(f"La moda es: {moda.mode} y aparece {moda.count} veces")
````

**Medidas de Tendencia Central Visualizadas**

Para visualizar mejor la distribución de los datos y las tres medidas de tendencia central (media, mediana y moda), podemos graficarlas juntas:

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Conjunto de datos
data = [85, 90, 78, 92, 88, 76, 95, 89, 85]

# Cálculo de medidas
media = np.mean(data)
mediana = np.median(data)
moda = stats.mode(data)

# Gráfico de barras
plt.figure(figsize=(10,6))
plt.bar(range(1, len(data)+1), sorted(data), color='lightblue', edgecolor='black')

# Añadir líneas para la media, mediana y moda
plt.axhline(y=media, color='r', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')
plt.axhline(y=mediana, color='g', linestyle='dashed', linewidth=2, label=f'Mediana = {mediana:.2f}')
plt.axhline(y=moda.mode, color='b', linestyle='dashed', linewidth=2, label=f'Moda = {moda.mode}')

# Etiquetas y leyenda
plt.title('Distribución de Calificaciones: Media, Mediana y Moda', fontsize=14)
plt.xlabel('Estudiantes', fontsize=12)
plt.ylabel('Calificaciones', fontsize=12)
plt.legend()
plt.grid(True)

# Mostrar gráfico
plt.show()
````

# 2. **Medidas de Dispersión**

### **Teoría General**

Las **medidas de dispersión** son fundamentales en el análisis de datos porque nos permiten entender **cuán dispersos o agrupados** están los valores de un conjunto de datos alrededor de una medida de tendencia central (como la media). Si bien las medidas de tendencia central (como la media, la mediana y la moda) nos dan una idea del valor "típico" o "central" de un conjunto de datos, las medidas de dispersión nos indican **cómo varían** esos datos.

En términos simples, las medidas de dispersión responden preguntas como:

- ¿Los valores están muy dispersos o concentrados alrededor de la media?
- ¿Existen diferencias significativas entre los valores más altos y más bajos?

Entre las principales medidas de dispersión se encuentran:

### 2.1 **Varianza**

- **Definición**: La **varianza** es una medida que refleja el promedio de las diferencias cuadráticas entre cada valor del conjunto de datos y la media. Es una forma de cuantificar **cuánto se alejan los datos de la media**. Cuanto mayor sea la varianza, más dispersos estarán los datos. Se mide en unidades cuadradas, lo que la hace difícil de interpretar directamente en comparación con los datos originales.
  
  Matemáticamente, la varianza se define como:
  
  $
  \text{Varianza} = \frac{\sum_{i=1}^{n} (x_i - \text{Media})^2}{n}
  $
  
  Donde:
  
  - $x_i$ son los valores individuales del conjunto de datos.
  - **Media** es el promedio de todos los valores.
  - **n** es el número total de datos.
- **Aplicación**:
  
  La varianza es útil cuando queremos cuantificar cuán "alejados" están los valores de la media, lo que nos proporciona una idea de la **variabilidad** dentro del conjunto de datos. Por ejemplo, en un grupo de estudiantes, si las calificaciones tienen una varianza baja, significa que la mayoría tiene calificaciones cercanas al promedio, pero si la varianza es alta, las calificaciones están muy dispersas (algunos estudiantes tienen notas mucho más altas o más bajas que la media).
  
- **Ventajas**:
  
  - Toma en cuenta **todos los valores** del conjunto de datos.
  - Es útil en muchos contextos estadísticos, especialmente en pruebas de hipótesis y modelos de regresión.
- **Limitaciones**:
  
  - Se mide en **unidades cuadradas**, lo que puede hacerla menos intuitiva para interpretar en comparación con los datos originales.

### 2.2 **Desviación Estándar**

- **Definición**: La **desviación estándar** es la raíz cuadrada de la varianza. Al ser la raíz cuadrada, devuelve las unidades de medida a las mismas unidades que los datos originales, lo que la hace más **fácil de interpretar**. Al igual que la varianza, mide la **dispersión** de los datos alrededor de la media.
  
  La desviación estándar se calcula como:
  
  $
  \text{Desviación Estándar} = \sqrt{\text{Varianza}} = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \text{Media})^2}{n}}
  $
  
  Es decir, es una medida que nos indica **en promedio cuánto se desvía cada valor de la media**.
  
- **Aplicación**:
  
  La desviación estándar se utiliza comúnmente en estudios de análisis de datos porque es fácil de interpretar en el contexto de los valores originales. Si la desviación estándar es **baja**, significa que la mayoría de los datos están cerca de la media; si es **alta**, indica que los datos están muy dispersos.
  
  Por ejemplo, en un análisis de salarios de empleados, si la desviación estándar es alta, algunos empleados ganan mucho más o mucho menos que el promedio.
  
- **Ventajas**:
  
  - Es más fácil de interpretar que la varianza, ya que está en las mismas unidades que los datos originales.
  - Útil para evaluar la **variabilidad** en comparación con la media en distribuciones simétricas.
- **Limitaciones**:
  
  - Al igual que la varianza, **es sensible a valores atípicos**. Un solo valor muy alto o bajo puede aumentar considerablemente la desviación estándar.

### 2.3 **Rango**

- **Definición**: El **rango** es la diferencia entre el valor más alto y el más bajo en un conjunto de datos. Es la medida de dispersión más sencilla, y nos da una idea rápida de la **amplitud** de los datos.
  
  Matemáticamente, se expresa como:
  
  $
  \text{Rango} = \text{Valor Máximo} - \text{Valor Mínimo}
  $
  
  Aunque el rango es fácil de calcular, **solo utiliza dos valores** del conjunto de datos (el mínimo y el máximo), por lo que no proporciona información sobre cómo se distribuyen los demás valores.
  
- **Aplicación**:
  
  El rango es útil cuando necesitamos una idea rápida del **intervalo** en el que caen todos los valores. Por ejemplo, si estamos analizando las temperaturas diarias de una ciudad, el rango nos indicará la diferencia entre el día más frío y el día más caluroso.
  
- **Ventajas**:
  
  - Es **fácil de calcular** y entender.
  - Nos da una idea rápida de la **amplitud** de los datos.
- **Limitaciones**:
  
  - Solo tiene en cuenta los **valores extremos**, ignorando completamente cómo se distribuyen el resto de los valores.
  - Es **muy sensible a valores atípicos**. Un valor extremadamente alto o bajo puede distorsionar el rango.

---

### **Cálculo de las Medidas de Dispersión en Python**

Para ilustrar mejor cómo calcular estas medidas de dispersión, utilicemos el siguiente conjunto de datos, que corresponde a las calificaciones de un grupo de estudiantes:

````python
import numpy as np

# Conjunto de datos: calificaciones de un grupo de estudiantes
data = [85, 90, 78, 92, 88, 76, 95, 89, 85]

# Cálculo de la varianza
varianza = np.var(data)
print(f"Varianza: {varianza:.2f}")

# Cálculo de la desviación estándar
desviacion_estandar = np.std(data)
print(f"Desviación Estándar: {desviacion_estandar:.2f}")

# Cálculo del rango
rango = np.max(data) - np.min(data)
print(f"Rango: {rango}")
````

**Interpretación de los Resultados**:

1. **Varianza**: Mide cuánto se alejan, en promedio, las calificaciones de la media. Si la varianza es alta, significa que las calificaciones están muy dispersas. En nuestro ejemplo, la varianza es 34.47, lo que indica que las calificaciones no están tan concentradas alrededor de la media.
  
2. **Desviación Estándar**: Al ser la raíz cuadrada de la varianza, nos indica en **promedio cuán lejos está cada valor de la media** en las mismas unidades que los datos originales (calificaciones). Por ejemplo, si la desviación estándar es 5.87, significa que, en promedio, las calificaciones varían en 5.87 puntos respecto a la media.
  
3. **Rango**: El rango mide la **amplitud total** de las calificaciones. Si el rango es 19 (la diferencia entre 95 y 76), indica que la calificación más alta es 19 puntos mayor que la más baja.
  
---

Las **medidas de dispersión** son esenciales para complementar las **medidas de tendencia central**. Mientras que la media, la mediana y la moda nos dicen cuál es el valor típico de un conjunto de datos, las medidas de dispersión nos dicen **cuán alejados o cercanos** están los datos entre sí y de ese valor típico. Con la varianza y la desviación estándar, obtenemos una idea precisa de la **variabilidad** del conjunto de datos, mientras que el rango nos proporciona una **visión rápida** de la amplitud de los valores.

# 3. **Cuantiles**

### **Teoría General**

Los **cuantiles** son valores que dividen un conjunto de datos ordenado en partes iguales, es decir, dividen los datos en diferentes segmentos o intervalos según la frecuencia acumulada. Son herramientas útiles para analizar la **distribución de los datos** y observar cómo se agrupan o se dispersan. Los cuantiles permiten identificar qué porcentaje de los datos se encuentra por debajo o por encima de un valor determinado.

Los tipos más comunes de cuantiles son:

### 3.1 **Cuartiles**

- **Definición**: Los **cuartiles** dividen un conjunto de datos en **cuatro partes iguales**, de modo que cada parte contenga el **25%** de los datos. Existen tres cuartiles principales que dividen el conjunto:
  
  - **Q1 (Primer Cuartil)**: El valor que separa el 25% inferior de los datos. Es decir, el **25% de los datos** son menores que este valor.
  - **Q2 (Segundo Cuartil o Mediana)**: Es la **mediana** del conjunto de datos y separa el 50% inferior del 50% superior.
  - **Q3 (Tercer Cuartil)**: El valor que separa el 75% inferior del 25% superior. El **75% de los datos** son menores que este valor.
  
  Matemáticamente, los cuartiles se calculan ordenando los datos y dividiéndolos en cuatro partes:
  
  - $ Q1 = \text{PERCENTIL}(25) $
  - $ Q2 = \text{PERCENTIL}(50) = \text{Mediana} $
  - $ Q3 = \text{PERCENTIL}(75) $
- **Aplicación**:
  
  Los cuartiles son útiles para identificar la **distribución de los datos** y entender cómo se agrupan los valores. Son fundamentales en el cálculo de medidas como el **rango intercuartílico**, que mide la dispersión de los datos sin verse afectado por valores extremos. Se aplican comúnmente en estudios de ingresos, análisis de puntuaciones, tiempos de espera, entre otros.
  
- **Ventajas**:
  
  - Proporcionan una forma clara de **dividir y analizar** un conjunto de datos.
  - Son útiles para estudiar la **distribución** sin verse afectados por valores atípicos o extremos.
- **Limitaciones**:
  
  - No ofrecen información sobre la dispersión de los datos dentro de cada intervalo, solo delimitan puntos específicos.

### 3.2 **Percentiles**

- **Definición**: Los **percentiles** dividen un conjunto de datos en **100 partes iguales**, de modo que cada percentil representa el valor por debajo del cual se encuentra un porcentaje específico de los datos. Los percentiles son utilizados con frecuencia para identificar rangos y comparar resultados.
  
  Por ejemplo:
  
  - **Percentil 25 (P25)**: El 25% de los datos son menores que este valor.
  - **Percentil 50 (P50)**: Es la mediana. El 50% de los datos son menores que este valor.
  - **Percentil 90 (P90)**: El 90% de los datos son menores que este valor.
  
  Matemáticamente, los percentiles se calculan ordenando los datos y localizando los puntos que dividen los datos en partes iguales:
  
  $
  \text{P}k = \text{Percentil} = \frac{k}{100} \cdot (n + 1) \text{, donde } k = \text{porcentaje}
  $
  
- **Aplicación**:
  
  Los percentiles son especialmente útiles para **comparaciones relativas** en distribuciones, como en exámenes o mediciones de rendimiento, donde se busca comparar cómo se desempeña un individuo en relación con un grupo más amplio. Son ampliamente utilizados en estadísticas de salud, como los percentiles de crecimiento infantil o en pruebas estandarizadas.
  
- **Ventajas**:
  
  - Proporcionan una herramienta precisa para **comparar posiciones** dentro de una distribución.
  - Son aplicables en diversos campos, desde la educación hasta la salud pública.
- **Limitaciones**:
  
  - Pueden ser **sensibles a valores extremos** si no se analiza su impacto en los percentiles.

---

### **Cálculo de Cuantiles Paso a Paso**

Ahora calcularemos y analizaremos los cuartiles y percentiles en Python, utilizando el mismo conjunto de datos de calificaciones de estudiantes que hemos empleado en las secciones anteriores:

````python
import numpy as np

# Conjunto de datos: calificaciones de un grupo de estudiantes
data = [85, 90, 78, 92, 88, 76, 95, 89, 85]

# Calcular los cuartiles
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # También conocido como la mediana
Q3 = np.percentile(data, 75)

# Calcular percentiles específicos
P90 = np.percentile(data, 90)

# Mostrar resultados
print(f"Q1 (Cuartil 1) = {Q1}")
print(f"Q2 (Mediana) = {Q2}")
print(f"Q3 (Cuartil 3) = {Q3}")
print(f"Percentil 90 = {P90}")
````

Paso a Paso del Cálculo de Cuantiles**:

1. **Cuartiles**:
  
  - **Q1**: Representa el valor por debajo del cual se encuentra el 25% de los datos.
  - **Q2 (Mediana)**: Representa el valor por debajo del cual se encuentra el 50% de los datos.
  - **Q3**: Representa el valor por debajo del cual se encuentra el 75% de los datos.
  
  En nuestro ejemplo, podríamos obtener que:
  
  - $ Q1 = 85 $
  - $ Q2 = 88 $ (la mediana)
  - $ Q3 = 90 $
2. **Percentiles**:
  
  - Por ejemplo, el **Percentil 90 (P90)** representa el valor por debajo del cual se encuentra el 90% de los datos.
  
  En nuestro ejemplo, $ P90 = 92.6 $, lo que significa que el 90% de los estudiantes obtuvo una calificación menor o igual a 92.8.

El valor del percentil 90 no toma un valor exacto de los datos porque es una estimación basada en la posición que ocupa dentro del conjunto ordenado de datos. En estadística, cuando calculamos percentiles (incluido el percentil 90), frecuentemente no coinciden exactamente con los valores específicos de los datos debido a la interpolación, especialmente cuando el percentil cae entre dos puntos de datos.

El percentil 90 indica el valor por debajo del cual el 90% de los datos se encuentran. Si este percentil no cae exactamente en un valor observado de los datos, se utiliza un método de interpolación para estimar el valor. Este es el motivo por el cual el percentil 90 puede aparecer como un número que no es un valor exacto de los datos.

Para entender mejor esto, veamos cómo se calcula típicamente un percentil cuando no coincide con un dato exacto:

1. **Ordenar los Datos**: Primero, los datos se ordenan en forma ascendente.
2. **Calcular la posición del percentil**: La posición del percentil $ p $ en un conjunto de datos con $ n $ elementos se calcula como:
   $
   \text{Posición} = \frac{p}{100} \times (n + 1)
   $
3. **Interpolación**: Si la posición calculada no es un número entero, se interpola entre los dos valores más cercanos.

Por ejemplo, si calculamos el percentil 90 para un conjunto de datos de 9 elementos:

- La posición teórica sería $ \frac{90}{100} \times (9 + 1) = 9.0 $, indicando el noveno valor.
- Si la posición fuese 8.5, esto implicaría que el percentil 90 está justo entre el octavo y el noveno valor de los datos ordenados.

Dado que en el gráfico proporcionado el percentil 90 es 92.6, este valor es una interpolación entre el noveno y décimo valor en el conjunto de datos ordenado (si hubiera un décimo valor). Si no hay un décimo valor, la interpolación puede depender de métodos específicos que pueden extender el rango de los datos existentes para estimar este percentil, o simplemente utilizar el valor más cercano que en este caso es el noveno valor.

Este método garantiza que los percentiles representen precisamente la distribución acumulativa de los datos, aún cuando no coincidan exactamente con los valores observados, lo cual es una práctica común en estadística para proporcionar una descripción detallada y precisa de la distribución de datos.
